```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=1000, centers=5, n_features=2,
                  cluster_std=1.2, random_state=17)
plt.scatter(X[:, 0], X[:, 1])

kM = KMeans(n_clusters=3, init='random', n_init=10,
            max_iter=350, tol=1e-4, random_state=None)
kM.fit(X)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], s=70)
plt.scatter(kM.cluster_centers_[:, 0], kM.cluster_centers_[:, 1],
            marker='*', s=200, color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')
plt.show()

print('Inertia: ', kM.inertia_)
```

```python
plt.figure(figsize=(15, 15))
for n_c in range(2, 8):
    k_means = KMeans(n_clusters=n_c)
    clusters = k_means.fit_predict(X)
    plt.subplot(3, 2, n_c-1)
    plt.scatter(X[:, 0], X[:, 1],  c=clusters)
    plt.title('n_clusters = {}'.format(n_c))
plt.show()
```
